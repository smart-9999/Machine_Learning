```python
import pandas as pd
import matplotlib as plt
import torch
import torch.nn.functional as F #Activiation functions
import torch.nn as nn  #addition of layer
```

```python
df = pd.read_csv('/kaggle/input/iriscsv/Iris.csv')
```

```python
df.head()
```

```python
from sklearn import preprocessing 
le = preprocessing.LabelEncoder() #converting to zeros,ones,twos
df['Species'] = le.fit_transform(df['Species'].values)
y = df['Species'].values #assign y
                               
                               
```

```python
y
```

```python
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler

X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size = 0.3)
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

```python
type(X_train)
```

```python
## Numpy -->TorchTensors #converts from numpy to pytorch tensors
from torch.autograd import Variable
X_train = Variable(torch.from_numpy(X_train)).float()
Y_train = torch.LongTensor(Y_train)
X_test = Variable(torch.from_numpy(X_test)).float()
Y_test = torch.LongTensor(Y_test)

```

```python
X_train.shape
```

```python
# Model #defining a model
# Input_dim :x_train.shape[1]
# hidden dim = 

class FeedForwardNN(nn.Module): #class objects
    
    def  __init__(self,input_dim,hidden,out):
    
        super(FeedForwardNN,self).__init__()
        self.relu = nn.ReLU() #activation function 
        self.layer1 = nn.Linear(input_dim,hidden)
        self.layer2 = nn.Linear(hidden,hidden)
        self.layer3 = nn.Linear(hidden,out)
    
    def forward(self,x): #flow 
        h1 = self.layer1(x)
        h2 = self.relu(h1)
        h3 = self.layer2(h2)
        h4 =self.relu(h3)
        out = self.relu(self.layer3(h4))
        
        return out 
    

    
```

```python
ffn = FeedForwardNN(input_dim = X_train.shape[1],hidden=32,out=3) 
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(ffn.parameters(),lr=0.1) 

```

```python
#ffn
```

```python
epochs =100
for epoch in range(epochs):
    y_pred =ffn(X_train)
    loss = loss_fn(y_pred,Y_train)
    #Gradient Descent 
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print(f"{epoch+1}/{epochs},Train Loss:{loss.item():.3f}")
    
    
    
```
